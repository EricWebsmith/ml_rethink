# Universal Filling for Isolation Forest

# Review of Isolation Forest

Isolation Forest is an anormaly detection algorithm proposed by Zhihua Zhou in 2008. The idea is to build an Isolation Tree and split randomly. Such that the anormalies are more likely to be singled out in early splitting and they will have a short path length to the root of the tree. The path length is used to determine the outlierness of a point. A bounch of Isolation Trees, usually 100, will form the Isolation Forest. 

One problem of Isolation Forest, as I will demonstrate later, is that it needs at least one point in certain area in order to have a short path. If there is no point, there is no path at all. And that area will be merged into other area. If they are merged into dense area, they will be considerd a dense area, too. 

# Universal Filling
Our solution to this problem is universal filling. The idea is to distribute additional points evenly across the whole map. Thus, we will have at least one point everywhere. Meanwhile the relative density stays the same. That is because if we give one more point to each place, the dense place is still dense and the sparse area is still sparse.

We can see that when sales is near zero. It is always normal. Think about it. How can you sell something that is near 0 dollars and make a lot of money? Do you think that is normal? Of cause not. So we have to fix this problem.

# Fix

To fix this problem, we add additional points. 

```python
X_fix = np.zeros((121, 2))
for i in range(10+1):
    for j in range(10+1):
        X_fix[i*11+j]=[i/10,j/10]
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(X[:,0], X[:,1], label='original data')
plt.scatter(X_fix[:,0], X_fix[:,1], label='universal filling')
plt.legend(prop=matplotlib.font_manager.FontProperties(size=15), loc='lower right')
plt.show()
```

```python
X_fixed=np.concatenate((X, X_fix), axis=0)
```

```python
outliers_fraction = 0.01
xx , yy = np.meshgrid(np.linspace(0, 1, 100), np.linspace(0, 1, 100))
clf = IForest(contamination=outliers_fraction,random_state=2)
clf.fit(X_fixed)
# predict raw anomaly score
scores_pred = clf.decision_function(X) * -1
        
# prediction of a datapoint category outlier or inlier
anomaly_labels=clf.predict(X)
n_inliers = len(y_pred) - np.count_nonzero(y_pred)
n_outliers = np.count_nonzero(y_pred == 1)
plt.figure(figsize=(8, 8))


         
print('OUTLIERS: ',n_outliers,'INLIERS: ',n_inliers)
        
# threshold value to consider a datapoint inlier or outlier
threshold = np.percentile(scores_pred, 100 * outliers_fraction)

anomaly_labels=scores_pred<threshold
        
# decision function calculates the raw anomaly score for every point
Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()]) * -1
Z = Z.reshape(xx.shape)
# fill blue map colormap from minimum anomaly score to threshold value
plt.contourf(xx, yy, Z, levels=np.linspace(Z.min(), threshold, 7),cmap=plt.cm.Blues_r)
        
# draw red contour line where anomaly score is equal to thresold
a = plt.contour(xx, yy, Z, levels=[threshold],linewidths=2, colors='red')
        
# fill orange contour lines where range of anomaly score is from threshold to maximum anomaly score
plt.contourf(xx, yy, Z, levels=[threshold, Z.max()],colors='orange')
b=plt.scatter(X[anomaly_labels==0, 0], X[anomaly_labels==0, 1], c='white', s=20, edgecolor='k', label='Normal')
c=plt.scatter(X[anomaly_labels==1, 0], X[anomaly_labels==1, 1], c='black', s=20, label='Anormaly')
plt.scatter(outliers_sales, outliers_profit, c='black',s=20, edgecolor='k') 
plt.axis('tight')
plt.legend([a.collections[0], b,c], ['learned decision function', 'inliers','outliers'],
           prop=matplotlib.font_manager.FontProperties(size=20),loc='lower right')
      
plt.xlim((0, 1))
plt.ylim((0, 1))
plt.title('Isolation Forest')
plt.show()
```

We see that with Universal Filling, we fixed the problem wrong labelling of the zero-point area. When We train anormaly detection algorithm,  we will have a large space of zero-point area. We should not label these area as the normal area. To the contrary, we should label them anormal. Thus, using Universal Filling is of crucial importance in your model.
